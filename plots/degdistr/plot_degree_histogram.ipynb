Read degree distribution from a text file vertex by vertex.

```python
# read degree dist file into an dict
def read_deg_dist(file):
    with open(file) as f:
        degree_sequence = {}
        deg = []
        cnt = []
        for line in f: # read rest of lines
            [k, v] = list(map(int, line.split()))
            deg.append(k)
            cnt.append(v)

    return deg, cnt
```


# Degree histogram


Draw degree histogram with matplotlib.
Random graph shown as inset


```python
# plot hist for degree distribution
def draw_deg_dist_hist(deg, cnt, file, width = 0.80, 
                       xlog = True, ylog = True,
                      scilimits_m = 0, scilimits_n = 0):
    font_size = 40
    fig, ax = plt.subplots(figsize=(5, 5))
    plt.bar(deg, cnt, width=width, color="0.5", edgecolor='black')

    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    #plt.title("Degree Histogram")
    #plt.ticklabel_format(style='sci', axis='both', scilimits=(scilimits_m, scilimits_n))
    #ax.set_ylabel("Count (in log scale)", fontsize = font_size)
    #ax.set_xlabel('Degree', fontsize = font_size)
    #ax.tick_params(labelsize = font_size/2)
    #ax.xaxis.get_offset_text().set_size(font_size/2)
    if xlog:
        ax.set_xscale('log')
    if ylog:
        ax.set_yscale('log')


    plt.show()
    fig.savefig(file + '.pdf', format = 'pdf', bbox_inches='tight')
    fig.savefig(file + '.eps', format = 'eps', bbox_inches='tight', dpi = fig.dpi)
    return
```

```python
file = 'dns/dns-001.mtx0dd'
deg, cnt = read_deg_dist(file)
draw_deg_dist_hist(deg, cnt, file, xlog=False)
```
